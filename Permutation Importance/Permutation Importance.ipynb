```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
```

```python
df=pd.read_csv("FIFA 2018 Statistics.csv")
```

```python
y=(df['Man of the Match']=='Yes')
```

```python
feature_names=[i for i in df.columns if df[i].dtype in ['int64']]
feature_names 
```

```python
x=df[feature_names]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1)
```

```python
my_model=RandomForestClassifier(n_estimators=100,random_state=0)
my_model.fit(x_train,y_train)
```

```python
# !pip install eli5
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm= PermutationImportance(my_model,random_state=1).fit(x_test,y_test)
eli5.show_weights(perm,feature_names=x_test.columns.to_list())
```

### Practice

```python
df=pd.read_csv("Texi train.csv")
df.shape
```

```python
df = df.query('pickup_latitude > 40.7 and pickup_latitude < 40.8 and ' +
                  'dropoff_latitude > 40.7 and dropoff_latitude < 40.8 and ' +
                  'pickup_longitude > -74 and pickup_longitude < -73.9 and ' +
                  'dropoff_longitude > -74 and dropoff_longitude < -73.9 and ' +
                  'fare_amount > 0'
                  )
```

```python
y=df.fare_amount
```

```python
x=df[['pickup_longitude',
                 'pickup_latitude',
                 'dropoff_longitude',
                 'dropoff_latitude',
                 'passenger_count']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1)
my_model = RandomForestRegressor(n_estimators=50, random_state=1)
my_model.fit(x_train,y_train)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

# Make a small change to the code below to use in this problem. 
perm = PermutationImportance(my_model, random_state=1).fit(x_test, y_test)
```

```python
eli5.show_weights(perm, feature_names = x_test.columns.tolist())
```

```python
# create new features
df['abs_lon_change'] = abs(df.dropoff_longitude - df.pickup_longitude)
df['abs_lat_change'] = abs(df.dropoff_latitude - df.pickup_latitude)

features_2  = ['pickup_longitude',
               'pickup_latitude',
               'dropoff_longitude',
               'dropoff_latitude',
               'abs_lat_change',
               'abs_lon_change']

x_new = df[features_2]
x_train_new,x_test_new , y_train_new, y_test_new = train_test_split(x_new, y, random_state=1)
second_model = RandomForestRegressor(n_estimators=30, random_state=1).fit(x_train_new, y_train_new)

# Create a PermutationImportance object on second_model and fit it to new_val_X and new_val_y
# Use a random_state of 1 for reproducible results that match the expected solution.
perm2 = PermutationImportance(second_model, random_state=1).fit(x_test_new,y_test_new)

# show the weights for the permutation importance you just calculated
eli5.show_weights(perm2, feature_names = x_test_new.columns.tolist())


```
